```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
```

```python
url ="https://www.flipkart.com/search?q=oppo+mobile&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_ps&as-pos=1&as-type=RECENT&suggestionId=oppo+mobile%7CMobiles&requestId=f6dff080-9ace-4a8c-a45e-950ed4d9d53d&as-searchtext=oppo"
url
```

```python
r = requests.get(url)
data = r.content
```

```python
soup = BeautifulSoup(data,'html.parser')
print(soup.prettify)
```

```python
#Find from where we did our Web-Scrapping
title = soup.title
title.string 
```

```python
model=soup.findAll('div',{'class':"_4rR01T"})
model
```

```python
productname=[]
for i in model:
    p=i.text
    productname.append(p)

print(productname)
```

```python
Specification=soup.find_all('div',class_="fMghEO")
Specification
```

```python
prices=soup.findAll('div',{'class':"_30jeq3 _1_WHN1"})
prices
```

```python
price=[]
for i in prices:
    pr=i.text
    price.append(pr)

print(price)
```

```python
r=soup.findAll('div',{'class':"_3LWZlK"})
rating=[]
for i in r:
    ra=i.text
    rating.append(ra)

print(rating)
```

```python
Ram=soup.find_all('div',{'class':'fMghEO'})
Ram
```

```python
Mobile_Ram=[]
for i in Ram:
    mem=i.find_all('li',class_="rgWa7D")
    m=mem[0].text     #fetching at index 0 
    Mobile_Ram.append(m)
print(Mobile_Ram)
```

```python
Display=soup.find_all('div',class_="fMghEO")
Display
```

```python
Mobile_Display=[]
for i in Display:
    dis=i.find_all('li',class_="rgWa7D")
    d=dis[1].text     #fetching at index 1
    Mobile_Display.append(d)
print(Mobile_Display)
```

```python
Camera=soup.find_all('div',{'class':'fMghEO'})
Camera
```

```python
Mobile_Camera=[]
for i in Display:
    cam=i.find_all('li',class_="rgWa7D")
    c=cam[2].text     #fetching at index 2 
    Mobile_Camera.append(c)
print(Mobile_Camera)
```

```python
Battery=soup.find_all('div',class_='fMghEO')
Battery
```

```python
Mobile_Battery=[]
for i in Battery:
    batt=i.find_all('li',class_="rgWa7D")
    ba=batt[3].text     #fetching at index 3
    Mobile_Battery.append(ba)
print(Mobile_Battery)
```

```python
Review=soup.findAll('span',{'class':"_2_R_DZ"})
Review
```

```python
Mobile_Review=[]
for i in Review:
    rew=i.text
    Mobile_Review.append(rew)
print(Mobile_Review)
```

```python
d={'productname':productname,'price':price,'rating':rating,'Mobile_Ram & Rom':Mobile_Ram,'Mobile_Display':Mobile_Display,'Mobile_Camera':Mobile_Camera,'Mobile_Battery':Mobile_Battery,'Review & Rating':Mobile_Review}
Product=pd.DataFrame(d)
Product
```

```python
Product.to_csv("mobiledata.csv")#transfer our scrap data to mobiledata file in csv format
```

```python
Product.info()
```

```python
Product['price'].value_counts().sort_values(ascending=False).plot(kind = 'bar',figsize=(20,10),color ='blue')
```

```python
battery=Product['Mobile_Battery'].value_counts()
battery
```

```python
plt.figure(figsize=(8,8))
battery.plot.pie(y=battery,autopct="%1.1f%%")
plt.text(-0.95,1.25,'PIE DIAGRAM SHOWING BATTERY CATEGORY',fontsize=20)
plt.legend()
```
